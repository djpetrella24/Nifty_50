```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study dataset
nse_india_path = "dataset/NSE_India.csv"

# Read dataset
nse_metadata = pd.read_csv(nse_india_path)

# Display the data table for preview
nse_metadata.head()
```

```python
# Remove commas in all columns of the dataset & check to see if it works. 
columns = ['Open', 'High', 'Low', 'LTP', 'Chng', '% Chng', 'Volume (lacs)', 'Turnover (crs.)', '52w H', '52w L']
nse_metadata[columns] = nse_metadata[columns].replace({',': ''}, regex=True).astype(float) 
nse_metadata.head()
```

```python
df = pd.DataFrame(nse_metadata)
df.head()
```

```python
opendf = df.get(['Symbol', 'Open'])
opendf.head()
```

## Summary Statistics

```python
# Calculate the mean and median for open values. 
opening_value = nse_metadata['Open']


mean_numpy = np.mean(opening_value)
print(f"The mean opening value of the Nifty 50 is {mean_numpy}")

median_numpy = np.median(opening_value)
print(f"The median opening value of the Nifty 50 is {median_numpy}")
```

```python
# Calculate the mean and median for high values. 
high_value = nse_metadata['High']


mean_numpy = np.mean(high_value)
print(f"The mean high value of the Nifty 50 is {mean_numpy}")

median_numpy = np.median(high_value)
print(f"The median high value of the Nifty 50 is {median_numpy}")
```

```python
# Calculate the mean and median for low values. 
low_value = nse_metadata['Low']


mean_numpy = np.mean(low_value)
print(f"The mean opening value of the Nifty 50 is {mean_numpy}")

median_numpy = np.median(low_value)
print(f"The median opening value of the Nifty 50 is {median_numpy}")
```
