```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study dataset
nse_india_path = "dataset/NSE_India.csv"

# Read dataset
nse_metadata = pd.read_csv(nse_india_path)

# Display the data table for preview
nse_metadata.head()
```

```python
# Remove commas in all columns of the dataset & check to see if it works. 
columns = ['Open', 'High', 'Low', 'LTP', 'Chng', '% Chng', 'Volume (lacs)', 'Turnover (crs.)', '52w H', '52w L']
nse_metadata[columns] = nse_metadata[columns].replace({',': ''}, regex=True)
nse_metadata.head()
```

## Summary Statistics

```python
opening_price = nse_metadata['Open']
opening_price.head()
```

```python
mean_numpy = np.mean(opening_price)
print(f"The mean open of the Nifty 50 is {mean_numpy}")
```
